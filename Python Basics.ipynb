```python
#Arithmetic Operators
a=20
b=5
#Addition
print(a+b)
#Subtraction
print(a-b)
#Multiplication
print(a*b)
#Float Division
print(a/b)
#Floor Division
print(a//b)
#Modulo Division
print(a%b)
#Power or Square
print(a**b)
```

```python
#Relational  Operators or Comparison Operators with Integer
a=20
b=2
print(a==b)
print(a!=b)
print(a<=b)
print(a>=b)
```

```python
#Relational  Operators or Comparison Operators with String
a="Prakash"
b="prakash"
print(a==b)
print(a!=b)
print(a>b)
print(a<b)
```

```python
#Augmented Assignment Operators
a=5
b=10
c=15
d=20
e=25
a+=2 #a=a+2
b-=2 #b=b-2
c*=2 #c=c*2
d/=2 #d=d/2
e%=2 #e=e%2
print(a)
print(b)
print(c)
print(d)
print(e)
```

```python
#Membership Operators
list=[2,4,6,8,3,5,7]
name='Prakash'
print('P' in name)
print('P' in name)
print('J' in name)
print('J' not in name)
print(8 in list)
print(8 not in list)
```

```python
#Logical Operators
num=10
print(num>5 and num<15)
print(num%2==0 or num%3==0)
print(not(num==0))
```

```python
#Bitwise Operators
x=10
y=4
print(x & y)
print(x | y)
print(x ^ y)
print(x<<y)
print(x>>y)
```
