```python
bmi=int(input("Enter the BMI Index:"))
if(bmi<18.5):
    print("Underweight")
elif(bmi<24.9):
    print("Normal Weight")
elif(bmi<29.9):
    print("Overweight")
elif(bmi<34.9):
    print("Very Overweight")
elif(bmi<39.9):
    print("Severely Overweight")
else:
    print("Morbidly Overweight")
```
